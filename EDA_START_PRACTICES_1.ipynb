**Step-4**

**Providing index**

```python
name=['swathi', 'sai', 'priya']
age=[22,23,24]
city=['tadipatri', 'nittur', 'yadiki']
cols=['NAMES', 'AGE', 'CITY']
ix=['A', 'B', 'C']
pd.DataFrame(zip(name,age,city),index=ix,columns=colls)
```

```python
name = ['Tarun', 'Venkat', 'Sahasra']
age = [20, 35, 12]
city = ['Hyd', 'Blr', 'Chn']
data = zip(name, age)
column=['Name', 'Age']
df=pd.DataFrame(data, index=city, columns=column)
df
```

```python
names=['krishna', 'swathi', 'veni']
age=[22,23,24]
height=[4,5,6]
col=['NAMES', 'AGE', 'HEIGHT']
ix=['A', 'B', 'C']
data=pd.DataFrame(zip(names,age,height),columns=col,index=ix)
data
```
